# GSEE: The simplest example

## Define the Hamiltonian H = Z

```python
from qiskit.quantum_info import SparsePauliOp
```

```python
H = SparsePauliOp("Z")
```

## Parameterized ansatz: a single Ry rotation is sufficient for 1 qubit

```python
from qiskit import QuantumCircuit
from qiskit.circuit import Parameter
```

```python
theta = Parameter("θ")
ansatz = QuantumCircuit(1)
ansatz.ry(theta, 0);
```

## Quiskit boiler plate, we need this to simulate
### Set up the estimator

```python
from qiskit.primitives import StatevectorEstimator
```

```python
estimator = StatevectorEstimator()
```

## Cost function: computes $\bra{\psi(\theta)} H \ket{\psi(\theta)}$

```python
def cost_func(params):
    job = estimator.run([(ansatz, H, [params])])
    return job.result()[0].data.evs[0]
```

## Optimize

```python
import numpy as np
from scipy.optimize import minimize
```

```python
x0 = np.array([0.5])  # initial guess
result = minimize(cost_func, x0, method="cobyla")
```

```python
print(f"Ground state energy: {result.fun:.6f}")  # Expected: -1.0
print(f"Optimal θ: {result.x[0]:.6f}")           # Expected: π
```
